```
import cv2
import dlib
import face_recognition
import os
import numpy as np
import pickle
```

```
# Face detection using dlib face detector
knownEmbeddings = []
knownNames = []
ImgPath = []
detector = dlib.get_frontal_face_detector()  #Dlib Face detector

predictor_68_point_model = face_recognition_models.pose_predictor_model_location()
pose_predictor_68_point = dlib.shape_predictor(predictor_68_point_model)
```

```
# loading of image from folder and feature extraction

from google.colab.patches import cv2_imshow
names = os.listdir('drive/My Drive/dataset')
total = 0
img_list=[]
for name in names :
  print(name)
  img_list=os.listdir('drive/My Drive/dataset' + '/' + name)
  print(img_list)
  cnt=0
  for img_path in img_list:
        pixels = cv2.imread('drive/My Drive/dataset' + '/' + name + '/' + img_path)
        #cv2_imshow(pixels)
        if pixels is None:
            continue
        if pixels.shape[0]<640 and pixels.shape[1]<480:    #discard face if its size is less than a particular value
            continue
        rgb = cv2.cvtColor(pixels, cv2.COLOR_BGR2RGB)
        boxes = face_recognition.face_locations(rgb)

        if len(boxes)==1: #consider only one face per image. If multiple faces present in the single image, discard them
            cnt += 1
            encodings = face_recognition.face_encodings(rgb, boxes,num_jitters=1)
            knownEmbeddings.append(encodings[0])
            knownNames.append(name)
            ImgPath.append(img_path)

        if cnt == 15:
            break
        print(name,cnt)
```

```
# dictionary from 3 list for labeling

data = {"embeddings": knownEmbeddings, "names": knownNames,"paths":ImgPath}
```

```
#write the dictionary in the pickle file
f = open('celeb_embedding.pickle', "wb")
f.write(pickle.dumps(data))
f.close()
```
